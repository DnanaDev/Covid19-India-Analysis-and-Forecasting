### **Points to consider :**
--------------
* Since the data is semi-structured, you should analyze the key-value pairs individually
  * Ex. Look at the key : extras. It will appear only when there is any extras conceded
  * Another one. Have a look at the key : wicket. It will appear only in the deliveries in which a wicket fell
* Your code should be able to take care of all such scenarios and create one big flat file

### **The structure of the csv is as follows**
* **delivery# :** delivery sequence
* **batsman :** name of the batsman who faced the delivery
* **bowler :** name of the bowler who bowled the delivery
* **non_striker :** name of the batsman at the non-striker end
* **runs_batsman :** runs made by the batsman in that delivery
* **runs_extras :** runs conceded by any extras thrown (ex. wide, no-ball)
* **extras_type :** the type of extras conceded
* **player_out :** the batsman who got out in this delivery
* **mode_of_dismissal :** the way in which the batsman got out (bowled, caught)
* **fielders_involved :** the list of fielders involved

*The structure mentioned above is best to my knowledge. Feel free to add any new fields in case you discover something new in the data*
