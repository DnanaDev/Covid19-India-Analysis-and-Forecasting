```python
from statsmodels.tsa.stattools import adfuller
```

```python
deaths1 = deaths.loc[:,'Lat':'3/23/20']
```

```python
print(" > Is the data stationary ?")
for i in deaths1.columns:
    dftest = adfuller(deaths1[i], autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

```python
recovered1 = recovered.loc[:,'Lat':'3/23/20']
```

```python
print(" > Is the data stationary ?")
for i in recovered1.columns:
    dftest = adfuller(recovered1[i], autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

```python
deaths_predicted = pd.DataFrame()
```

```python
for i in deaths_transposed.columns:
    mod = sm.tsa.statespace.SARIMAX(deaths_transposed[i].astype('int64'), trend='n', order=(0,1,0), seasonal_order=(1,1,1,12))
    results = mod.fit()
    print(i)
    print(results.predict('3/21/20', '3/23/20', dynamic=False))
    deaths_predicted[i] = results.predict('3/24/20', '3/30/20', dynamic=False)
```
